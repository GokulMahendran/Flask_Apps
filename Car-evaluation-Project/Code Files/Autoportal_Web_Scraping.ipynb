```
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
import numpy as np
import json
from google.colab import files
```

```
res=[]
content=[]
base_url ="https://autoportal.com/usedcars/finder/"
for i in range(1,478):
  print(i)
  response = get(base_url+"page-"+str(i))
  soup=BeautifulSoup(response.text,"html.parser")
  content.extend(soup.find_all("span",attrs={"class":"contact-button"}))
for j in content:
  res.append(json.loads(j.get("data-announcement-raw")))
```

```
links=[]
base_url ="https://autoportal.com/usedcars/finder/"
for i in range(1,478):
  print(i)
  response = get(base_url+"page-"+str(i))
  soup=BeautifulSoup(response.text,"html.parser")
  a=soup.find_all("a",attrs={"class":"link"})
  if a is None:
    continue
  else:
    links.extend([i.get("href") for i in a])
```

```
def url_find(i): 
  response = get("https://autoportal.com/"+i)
  soup=BeautifulSoup(response.text,"html.parser")
  a=soup.find_all("div",attrs={"class":"table-in title"})
  d={}
  for i in range(len(a)):
    f=a[i].text.split(":")
    d.update({f[0].strip():f[1].strip()})
  if "Transmission" in d.keys():
    return d["Transmission"]
  else:
    return np.nan
```
